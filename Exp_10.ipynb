## Resize the image

## Flatten the images

```python
len_of_images=len(Priyanka_chopra_gray)
```

```python
image_size=(160,160)
```

```python
flatten_size=image_size[0]*image_size[1]
```

```python
flatten_size
```

```python
for i in range(len_of_images):
    Akshay_kumar_gray[i] = np.ndarray.flatten(Akshay_kumar_gray[i]).reshape(flatten_size)
    Alia_bhatt_gray[i] = np.ndarray.flatten(Alia_bhatt_gray[i]).reshape(flatten_size)
    Anushak_sharma_gray[i] = np.ndarray.flatten(Anushak_sharma_gray[i]).reshape(flatten_size)
    Priyanka_chopra_gray[i] = np.ndarray.flatten(Priyanka_chopra_gray[i]).reshape(flatten_size)
```

```python
Akshay_kumar_gray = np.dstack(Akshay_kumar_gray) 
Alia_bhatt_gray = np.dstack(Alia_bhatt_gray)
Anushak_sharma_gray = np.dstack(Anushak_sharma_gray)
Priyanka_chopra_gray = np.dstack(Priyanka_chopra_gray)
```

```python
Priyanka_chopra_gray
```

```python
Priyanka_chopra_gray.shape
```

```python
Akshay_kumar_gray = np.rollaxis(Akshay_kumar_gray, axis=2, start=0) 
Alia_bhatt_gray = np.rollaxis(Alia_bhatt_gray, axis=2, start=0)
Anushak_sharma_gray = np.rollaxis(Anushak_sharma_gray, axis=2, start=0)
Priyanka_chopra_gray = np.rollaxis(Priyanka_chopra_gray, axis=2, start=0)
```

```python
Akshay_kumar_gray = Akshay_kumar_gray.reshape(len_of_images,flatten_size) 
Alia_bhatt_gray = Alia_bhatt_gray.reshape(len_of_images,flatten_size)
Anushak_sharma_gray = Anushak_sharma_gray.reshape(len_of_images,flatten_size)
Priyanka_chopra_gray = Priyanka_chopra_gray.reshape(len_of_images,flatten_size)
```
